```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("car.csv")
print(df.head())
df.info()

duplicates = df.duplicated().sum()
print(f"Number of duplicates: {duplicates}")
df = df.drop_duplicates()
print("\nMissing values per column:")
print(df.isnull().sum())
df.dropna(subset=['TotalCharges'], inplace=True)
print(f"\nنسبة المفقود في TotalCharges: {df['TotalCharges'].isnull().sum() / len(df) * 100:.2f}%")

numeric_columns = ['tenure', 'MonthlyCharges', 'TotalCharges']

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

for column_to_check in numeric_columns:
    print(f"\n--- Statistics for {column_to_check} ---")
    print(df[column_to_check].describe())

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    sns.histplot(df[column_to_check], kde=True, ax=axes[0])
    axes[0].set_title(f"Histogram - {column_to_check} (Shape)")

    sns.boxplot(x=df[column_to_check], ax=axes[1])  # ✅ أضفنا ax=axes[1]
    axes[1].set_title(f"Boxplot - {column_to_check} (Find Outliers)")

    plt.suptitle(f"Analysis of {column_to_check}", fontsize=14)
    plt.tight_layout()
    plt.show()
categorical_columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
                        'PhoneService', 'MultipleLines', 'InternetService',
                        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                        'TechSupport', 'StreamingTV', 'StreamingMovies',
                        'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn']

for category_col in categorical_columns:
    print(f"\nCounts for {category_col}:")
    print(df[category_col].value_counts())


    plt.figure(figsize=(8, 4))
    sns.countplot(x=df[category_col], order=df[category_col].value_counts().index)
    plt.title(f"Number of Customers in each {category_col}")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
numeric_pairs = [
    ('tenure', 'MonthlyCharges'),
    ('tenure', 'TotalCharges'),
    ('MonthlyCharges', 'TotalCharges')
]

for x_col, y_col in numeric_pairs:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=x_col, y=y_col, data=df,
                    hue='Churn',
                    s=50)
    plt.title(f"Relationship: {x_col} vs. {y_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()


plt.figure(figsize=(8, 6))
sns.heatmap(df[['tenure', 'MonthlyCharges', 'TotalCharges']].corr(),
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            vmin=-1, vmax=1)
plt.title("Correlation Heatmap - Numeric Columns")
plt.tight_layout()
plt.show()
```
